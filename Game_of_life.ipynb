## Game Of Life

```python
import numpy as np
import matplotlib.pyplot as plt

m,n = 30,30
a = np.round(np.random.rand(m,n))
maxit = 20
```

```python
def lnc(i, j, a): #living neighbours
    x = i
    y = j
    neighbours = 0

    """Return neighbours of given cell."""
    cellsToCheck = [(x-1, y-1), (x, y-1),  (x+1, y-1), 
                    (x-1, y),               (x+1, y), 
                    (x-1, y+1), (x, y+1), (x+1, y+1),]
    for y,x in cellsToCheck:
        if x >= 0 and x< np.size(a[0]) and y >= 0 and y < np.size(a[1]):
            # neighbour is alive
            neighbours += a[x, y]
    return int(neighbours)
```

```python
from IPython.display import clear_output, display
import time

fig = plt.figure()
ax = fig.add_subplot()
```

```python
aneu = np.copy(a)

for its in range(maxit):
    for i in range(m):
        for j in range(n):
            k = lnc(i, j, a)
            if a[i, j] == 0 and k == 3:
                aneu[i, j] = 1
            else:
                if k<2 or k>3:
                    aneu[i, j] = 0
    if (a == aneu).all() or (aneu == np.zeros((m,n))).all():
        break
    else:
        ax.spy(aneu)
        display(fig)
        clear_output(wait=True)
        a = np.copy(aneu)

ax.spy(aneu)
display(fig)
```
